```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str[1].str.split(')').str[0] == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
## Remove USA in event name 

df2['Event name'] = df2['Event name'].str.split('(').str[0]
```

```python
df2.head()
```

```python
## Cleanup athelete age 

df2['athelete_age'] = 2020 - df['Athlete year of birth']
```

```python
df2.head(5)
```

```python
## Remove h from athelete performance 

df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str[0]
```

```python
df2.head(5)
```

```python
# Drop columns - Athelete club, country, athelete year of birth, athelete age category

df2.drop(columns = ['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1, inplace = True)
```

```python
df2.head()
```

```python
## Cleanup Null Values 

df2.isna().sum()
```

```python
df2[df2['athelete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# Check duplicate values 

df2[df2.duplicated() == True]
```

```python
## Reset Index 

df2.reset_index(drop = True)
```

```python
# Fix datatypes 

df2.dtypes
```

```python
df2['athelete_age'] = df2['athelete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.info()
```

```python
df2.head()
```

```python
## Rename columns 

df2 = df2.rename(columns = {
    'Year of event' : 'year',
    'Event dates' : 'race_date',
    'Event name' : 'race_name',
    'Event distance/length' : 'race_length',
    'Event number of finishers' : 'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed' : 'athlete_average_speed',
    'Athlete ID' : 'athlete_id',
    'athelete_age' : 'athlete_age' 
})
```

```python
df2.head()
```

```python
# reorder columns 

df3 = df2[['race_date', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
# What age groups are best in the 50km race
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).round(2).sort_values('athlete_age', ascending = True).query('count > 19')
```

```python
# ## Seasons for the data -> Slower in summer than winter ?

# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2

# split between two decimals
```

```python
df3['race_month'] = df3['race_date'].str.split('.').str[1].astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x : 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'sum', 'count']).sort_values('mean', ascending = False)
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'sum', 'count']).sort_values('mean', ascending = False)
```
